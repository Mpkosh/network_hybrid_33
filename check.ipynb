```python
def calculate_incidence(E, S):
    '''
    Calculates newly infected cases using SEIR compatrments arrays.
    '''
    daily_incidence = [0 if index == 0 else ((E[index-1] - E[index]) -
                              (S[index] - S[index-1])) for index in range(len(S))]
    
    
    return daily_incidence
```
